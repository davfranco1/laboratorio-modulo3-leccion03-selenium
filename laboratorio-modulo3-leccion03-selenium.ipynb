```python
def palabra_aleatoria(lista):
    return random.choices(lista, k=2)
```

```python
lista_final = palabra_aleatoria(lista_municipios)
lista_final
```

```python
def crear_df(sopa, municipio, mes):
    lista_df = []
    for tabla in sopa:
        rows = tabla.findChildren("tr", recursive=False)
        df = pd.DataFrame([row.text for row in rows])
        lista_df.append(df)
    
    df_completo = pd.concat(lista_df, axis=1)
    df_completo.columns = ["Día", "Temp.", "DP", "Hum.", "WS", "Press.", "Prec."]
    df_final = df_completo.iloc[1:]
    
    meses = {"-1": "Enero", "-2": "Febrero", "-3": "Marzo", "-4": "Abril", "-5": "Mayo", "-6": "Junio", 
                  "-7": "Julio", "-8": "Agosto", "-9": "Septiembre", "10": "Octubre", "11": "Noviembre", "12": "Diciembre"}
    df_final["Mes"]=mes
    df_final["Mes"] = df_final["Mes"].replace(meses)
    df_final["Municipio"] = municipio
    
    for col in ["Temp.", "DP", "Hum.", "WS", "Press."]:
        cols = ["Max", " Avg", " Min"]
        df_final[[col + c for c in cols]] = df_final[col].str.split(expand=True)
        df_final.drop(columns=col, inplace=True)
    
    df_final["Prec."] = df_final.pop("Prec.")
    
    return df_final
```

```python
driver = webdriver.Chrome()
url_wunder = "https://www.wunderground.com/history"
driver.get(url_wunder)  # Navegar a la URL de Wunderground
driver.maximize_window()  # Maximizar la ventana del navegador
time.sleep(2)

try:
    iframe = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="sp_message_iframe_1165301"]')))
    driver.switch_to.frame(iframe)  # Cambiar al iframe de mensaje
    driver.find_element(By.CSS_SELECTOR, "#notice > div.message-component.message-row.cta-buttons-container > div.message-component.message-column.cta-button-column.reject-column > button").click()  # Cerrar el mensaje de cookies
    driver.switch_to.default_content()  # Volver al contenido principal
except Exception as e:
    print(f"Error handling cookies: {e}")  # Imprimir error en el manejo de cookies
time.sleep(2) 

dfs_municipios = []  # Lista para almacenar los dataframes de los municipios

for municipio in lista_final:
    try:
        driver.find_element(By.CSS_SELECTOR, "#historySearch").send_keys(municipio, Keys.TAB)  # Introducir el nombre del municipio
        time.sleep(2)  
        driver.find_element(By.CSS_SELECTOR, "#dateSubmit").click()  # Hacer clic en el botón de enviar
        time.sleep(2)
        driver.find_element(By.CSS_SELECTOR, "#dateSubmit").click()  # Hacer clic en el botón de enviar de nuevo
        time.sleep(2)
        driver.find_element(By.CSS_SELECTOR, "#inner-content > div.region-content-main > div.row > div:nth-child(1) > div:nth-child(1) > div > lib-link-selector > div > div > div > a:nth-child(3)").click()  # Hacer clic en el enlace de selector de enlaces
        time.sleep(2) 
    except Exception as e:
        print(f"Error navigating URL for {municipio}: {e}")  # Imprimir error al navegar la URL para el municipio
        continue
    
    url_actual = driver.current_url  # Obtener la URL actual
    lista_urls = [url_actual[:-2] + str(mes) for mes in range(1, 12)]  # Lista de URLs para los primeros 10 meses
    lista_tablas = []  # Lista para almacenar las tablas de datos

    for url in lista_urls:
        try:
            driver.get(url)  # Navegar a la URL
            mes_url = url[-2:]  # Obtener el mes de la URL
            time.sleep(2)
            html_table_page = driver.page_source  # Obtener el código fuente de la página
            sopa = BeautifulSoup(html_table_page, "html.parser")  # Analizar el código fuente con BeautifulSoup
            tabla = sopa.findAll("table", {"aria-labelledby": "Days data"})  # Encontrar la tabla de datos
            if not tabla:
                print(f"No se encontró tabla para la URL: {url}")  # Imprimir mensaje si no se encontró tabla
                continue
            
            tabla_pagina = crear_df(tabla, municipio, mes_url)  # Crear dataframe de la tabla
            if tabla_pagina is not None:
                lista_tablas.append(tabla_pagina)  # Agregar la tabla a la lista de tablas
            else:
                print(f"No se ha encontrado la tabla del municipio {municipio} para {mes_url}")  # Imprimir mensaje si no se encontró la tabla del municipio para el mes
        except Exception as e:
            print(f"Error procesando la URL {url}, motivo: {e}")  # Imprimir error al procesar la URL

    if lista_tablas:
        df_unido = pd.concat(lista_tablas, ignore_index=True)  # Unir todas las tablas en un dataframe
        dfs_municipios.append(df_unido)  # Agregar el dataframe unido a la lista de dataframes de municipios
        driver.get(url_wunder)  # Volver a la URL principal de Wunderground
    else:
        print(f"No se encontraron tablas de datos de {municipio}")  # Imprimir mensaje si no se encontraron tablas de datos del municipio
        driver.get(url_wunder)  # Volver a la URL principal de Wunderground

driver.quit()  # Cerrar el navegador
df_final = pd.concat(dfs_municipios, ignore_index=True)  # Unir todos los dataframes de municipios en uno final
```
